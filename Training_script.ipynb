```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, GroupKFold, GridSearchCV, StratifiedKFold
from sklearn.metrics import (accuracy_score, precision_score, classification_report, recall_score, 
                             f1_score, roc_auc_score, roc_curve, ConfusionMatrixDisplay, 
                             confusion_matrix, precision_recall_curve, make_scorer)
from sklearn.preprocessing import StandardScaler # Not actively used, but imported
import xgboost as xgb
import time
from collections import Counter
import shap 
from sklearn.inspection import PartialDependenceDisplay 

# ================ 0. Helper Function Definitions ================
def evaluate_model_metrics(model, X_test_data, y_test_data, model_name="Model"):
    y_pred = model.predict(X_test_data)
    metrics = {}
    y_test_data_int = y_test_data.astype(int)
    y_pred_proba = None # Initialize

    if len(np.unique(y_test_data_int)) < 2:
        metrics['accuracy'] = accuracy_score(y_test_data_int, y_pred)
        metrics['precision'] = precision_score(y_test_data_int, y_pred, zero_division=0, average='binary' if hasattr(model, 'classes_') and len(model.classes_)==2 else 'weighted', pos_label=1 if hasattr(model, 'classes_') and len(model.classes_)==2 else 0)
        metrics['recall'] = recall_score(y_test_data_int, y_pred, zero_division=0, average='binary' if hasattr(model, 'classes_') and len(model.classes_)==2 else 'weighted', pos_label=1 if hasattr(model, 'classes_') and len(model.classes_)==2 else 0)
        metrics['f1'] = f1_score(y_test_data_int, y_pred, zero_division=0, average='binary' if hasattr(model, 'classes_') and len(model.classes_)==2 else 'weighted', pos_label=1 if hasattr(model, 'classes_') and len(model.classes_)==2 else 0)
        metrics['roc_auc'] = None
        metrics['report'] = f"Only one class ({np.unique(y_test_data_int)[0]}) in y_test."
    else:
        metrics['accuracy'] = accuracy_score(y_test_data_int, y_pred)
        metrics['precision'] = precision_score(y_test_data_int, y_pred, zero_division=0, pos_label=1)
        metrics['recall'] = recall_score(y_test_data_int, y_pred, zero_division=0, pos_label=1)
        metrics['f1'] = f1_score(y_test_data_int, y_pred, zero_division=0, pos_label=1)
        try:
            y_pred_proba = model.predict_proba(X_test_data)[:, 1]
            metrics['roc_auc'] = roc_auc_score(y_test_data_int, y_pred_proba)
        except Exception: 
            metrics['roc_auc'] = None
            y_pred_proba = None # Ensure it's None if AUC calculation failed
        metrics['report'] = classification_report(y_test_data_int, y_pred, zero_division=0)
    return metrics, y_pred_proba # Return probabilities for ROC curve plotting

# ================ 1. Load and Prepare Data ================
print("== 1. Loading and Preparing Data ==")
try:
    df_original = pd.read_csv(r'C:\Tugas\Sem 8\TA\Data\Merged_Training_Data_Ver10.csv') 
    df = df_original.copy() 
except FileNotFoundError:
    print("FATAL ERROR: CSV file not found. Please check the path.")
    exit()

print("\n--- Processing 'Area' column based on 'Label', 'Field', and 'Field_2' ---")
if 'Label' in df.columns:
    df['Label_numeric'] = pd.to_numeric(df['Label'], errors='coerce')
else:
    df['Label_numeric'] = pd.NA 
df['Area'] = pd.NA 
field_col_exists = 'Field' in df.columns; field2_col_exists = 'Field_2' in df.columns
area_source_field = df['Field'].copy() if field_col_exists else pd.Series(pd.NA, index=df.index)
area_source_field2 = df['Field_2'].copy() if field2_col_exists else pd.Series(pd.NA, index=df.index)
mask_label1 = df['Label_numeric'] == 1
if field2_col_exists: df.loc[mask_label1, 'Area'] = area_source_field2[mask_label1]
mask_label0 = df['Label_numeric'] == 0
if field_col_exists: df.loc[mask_label0, 'Area'] = area_source_field[mask_label0]
mask_area_still_na = df['Area'].isna()
if mask_area_still_na.any():
    if field2_col_exists: df.loc[mask_area_still_na, 'Area'] = df.loc[mask_area_still_na, 'Area'].fillna(area_source_field2[mask_area_still_na]); mask_area_still_na = df['Area'].isna() 
    if mask_area_still_na.any() and field_col_exists: df.loc[mask_area_still_na, 'Area'] = df.loc[mask_area_still_na, 'Area'].fillna(area_source_field[mask_area_still_na])
df = df.drop(columns=['Field', 'Field_2', 'Label_numeric'], errors='ignore')
if 'Area' in df.columns: print(f"Final 'Area' column. Nulls: {df['Area'].isnull().sum()}")
else: print("CRITICAL: 'Area' column not constructed.")

feature_cols = ['LapseRate', 'NDWI', 'NDVI', 'Z', 'SMI','Linedense','TCI'] 
print(f"\nUsing feature columns: {feature_cols}")
existing_feature_cols_in_df = [col for col in feature_cols if col in df.columns]
if len(existing_feature_cols_in_df) != len(feature_cols): feature_cols = existing_feature_cols_in_df 
if feature_cols: df_cleaned = df.dropna(subset=feature_cols).copy()
else: print("No valid features for NaNs drop."); df_cleaned = df.copy()
print(f"Shape after initial feature NaNs drop: {df.shape} -> {df_cleaned.shape}")

training_data = pd.DataFrame(); prediction_data = pd.DataFrame()
if 'Label' in df_cleaned.columns:
    df_cleaned.loc[:, 'Label_for_split'] = pd.to_numeric(df_original.loc[df_cleaned.index, 'Label'], errors='coerce')
    known_mask = df_cleaned['Label_for_split'].notna()
    training_data = df_cleaned[known_mask].copy(); prediction_data = df_cleaned[~known_mask].copy() 
    training_data.drop(columns=['Label_for_split'], inplace=True, errors='ignore'); prediction_data.drop(columns=['Label_for_split'], inplace=True, errors='ignore')
    if not training_data.empty:
        if 'Label' in training_data.columns:
            training_data.loc[:, 'Label'] = pd.to_numeric(training_data['Label'], errors='coerce').fillna(-999).astype(int)
            training_data = training_data[training_data['Label'] != -999]
        if 'Area' in training_data.columns and 'Label' in training_data.columns and not training_data.empty:
            print("\n--- Class Distribution per Area (Training Data) ---")
            try: area_class_counts = training_data.groupby('Area')['Label'].value_counts().unstack(fill_value=0); print(area_class_counts); print(f"Unique areas for CV: {training_data['Area'].nunique()}")
            except Exception as e: print(f"Area class counts error: {e}")
    print(f"Shapes: Training {training_data.shape}, Prediction {prediction_data.shape}")
else: print("\nCRITICAL: 'Label' column not in df_cleaned.")

if not training_data.empty and feature_cols:
    print("\n--- Converting features to numeric (training_data) ---")
    for col in feature_cols:
        if col in training_data.columns: training_data.loc[:, col] = pd.to_numeric(training_data[col], errors='coerce')
    initial_rows = len(training_data); training_data.dropna(subset=feature_cols, inplace=True)
    if initial_rows > len(training_data): print(f"Training: {initial_rows - len(training_data)} rows dropped post-numeric conversion.")
if not prediction_data.empty and feature_cols:
    print("\n--- Converting features to numeric (prediction_data) ---")
    for col in feature_cols:
        if col in prediction_data.columns: prediction_data.loc[:, col] = pd.to_numeric(prediction_data[col], errors='coerce')
    initial_rows = len(prediction_data); prediction_data.dropna(subset=feature_cols, inplace=True)
    if initial_rows > len(prediction_data): print(f"Prediction: {initial_rows - len(prediction_data)} rows dropped post-numeric conversion.")

if not training_data.empty:
    if 'Area' in training_data.columns:
        initial_rows = len(training_data); training_data.dropna(subset=['Area'], inplace=True)
        if initial_rows > len(training_data): print(f"Training: {initial_rows - len(training_data)} rows dropped (NaN in 'Area'). Final CV shape: {training_data.shape}")
    elif not training_data.empty : print("CRITICAL: 'Area' col missing from final training_data.")
        
X_full_labeled, y_full_labeled, groups_full_labeled = np.array([]), np.array([]), np.array([])
if not training_data.empty and 'Label' in training_data.columns and 'Area' in training_data.columns and feature_cols and all(col in training_data.columns for col in feature_cols): 
    print("\n--- Dtypes of features in training_data before X_full_labeled ---"); training_data[feature_cols].info(verbose=False)
    X_full_labeled = training_data[feature_cols].values; y_full_labeled = training_data['Label'].values.astype(int); groups_full_labeled = training_data['Area'].values 
    if pd.isnull(groups_full_labeled).any(): print("CRITICAL ERROR: NaN in 'groups_full_labeled'.")
    print(f"\nCV Data: X {X_full_labeled.shape}, y {y_full_labeled.shape}, Groups unique: {pd.Series(groups_full_labeled).nunique()}")
    if pd.Series(groups_full_labeled).nunique() < 2 and len(training_data) > 0: print("CRITICAL: <2 unique areas for GroupKFold.")
else: print("Training data empty or key cols missing for CV setup.")
X_predict_unlabeled = np.array([])
if not prediction_data.empty and feature_cols and all(col in prediction_data.columns for col in feature_cols):
    X_predict_unlabeled = prediction_data[feature_cols].values
else: print("No unlabeled data/features for prediction.")

# ================ 2. Nested Cross-Validation and Hyperparameter Tuning ================
print("\n== 2. Nested Cross-Validation and Hyperparameter Tuning ==")
final_best_model = None; best_hyperparameters_overall = None; outer_cv_performances = []
if X_full_labeled.size > 0 and y_full_labeled.size > 0 and groups_full_labeled.size > 0:
    num_unique_areas_for_cv = pd.Series(groups_full_labeled).nunique()
    N_OUTER_SPLITS = 12
    if num_unique_areas_for_cv < N_OUTER_SPLITS : N_OUTER_SPLITS = num_unique_areas_for_cv
    if N_OUTER_SPLITS < 2: print(f"CRITICAL: Not enough unique areas for {N_OUTER_SPLITS} outer GroupKFold. Skipping CV.")
    else:
        print(f"Using N_OUTER_SPLITS = {N_OUTER_SPLITS} for GroupKFold."); outer_cv = GroupKFold(n_splits=N_OUTER_SPLITS)
        N_INNER_SPLITS = 2; param_grid_xgb = {'max_depth': [3,5,7], 'learning_rate': [0.05,0.1,0.2], 'n_estimators': [50,100,150,200,250,300]} 
        fold_count = 0; all_best_params_from_gs = []; all_feature_importances_outer = [] 
        for train_outer_idx, test_outer_idx in outer_cv.split(X_full_labeled, y_full_labeled, groups_full_labeled):
            fold_count += 1; print(f"\n--- Outer Fold {fold_count}/{N_OUTER_SPLITS} ---")
            X_train_outer,X_test_outer=X_full_labeled[train_outer_idx],X_full_labeled[test_outer_idx]; y_train_outer,y_test_outer=y_full_labeled[train_outer_idx],y_full_labeled[test_outer_idx]; groups_train_outer=groups_full_labeled[train_outer_idx] 
            if not (len(X_train_outer)>0 and len(X_test_outer)>0 and len(np.unique(y_train_outer))>=2):
                print(f"Skipping Fold {fold_count}: Empty train/test or single class in train ({np.unique(y_train_outer)})."); outer_cv_performances.append(None);all_best_params_from_gs.append(None);continue
            inner_cv_strategy=None; num_unique_groups_in_outer_train=pd.Series(groups_train_outer).nunique(); actual_inner_splits=min(N_INNER_SPLITS,num_unique_groups_in_outer_train if num_unique_groups_in_outer_train>1 else 2)
            if num_unique_groups_in_outer_train>=actual_inner_splits and actual_inner_splits>1: inner_cv_strategy=GroupKFold(n_splits=actual_inner_splits)
            else:
                min_samples_per_class=min(np.bincount(y_train_outer)) if len(np.unique(y_train_outer))>1 else 0; actual_inner_splits_strat=min(N_INNER_SPLITS,max(2,min_samples_per_class) if min_samples_per_class>1 else 2)
                if actual_inner_splits_strat<2: actual_inner_splits_strat=2; inner_cv_strategy=StratifiedKFold(n_splits=actual_inner_splits_strat,shuffle=True,random_state=fold_count)
            neg_co,pos_co=np.sum(y_train_outer==0),np.sum(y_train_outer==1);scale_w=neg_co/pos_co if pos_co*neg_co>0 else 1.0
            xgb_est=xgb.XGBClassifier(objective='binary:logistic',use_label_encoder=False,eval_metric='logloss',random_state=fold_count,scale_pos_weight=scale_w)
            scorer=make_scorer(precision_score,pos_label=1,zero_division=0)
            gs=GridSearchCV(estimator=xgb_est,param_grid=param_grid_xgb,scoring=scorer,cv=inner_cv_strategy,verbose=0,n_jobs=-1,error_score=0.0)
            try:
                fit_params={'groups':groups_train_outer} if isinstance(inner_cv_strategy,GroupKFold) else {}; gs.fit(X_train_outer,y_train_outer,**fit_params)
                all_best_params_from_gs.append(gs.best_params_);best_model_fold=gs.best_estimator_
                imp_gain=best_model_fold.get_booster().get_score(importance_type='gain');fold_imp_df=pd.DataFrame(list(imp_gain.items()),columns=['feature',f'imp_fold_{fold_count}']);all_feature_importances_outer.append(fold_imp_df.set_index('feature'))
                fold_perf,_ = evaluate_model_metrics(best_model_fold,X_test_outer,y_test_outer,f"Fold {fold_count}") # Get probs
                outer_cv_performances.append(fold_perf)
                print(f"    Fold {fold_count} Test: Precision: {fold_perf.get('precision','N/A'):.4f}, AUC: {fold_perf.get('roc_auc','N/A') if fold_perf.get('roc_auc') else 'N/A'}")
            except Exception as e: print(f"    Error GridSearch Fold {fold_count}: {e}");outer_cv_performances.append(None);all_best_params_from_gs.append(None)
        if outer_cv_performances:
            print("\n--- Aggregated Outer CV Perf ---");avg_mets={};valid_fold_perfs=[p for p in outer_cv_performances if p and isinstance(p,dict)]
            if valid_fold_perfs and valid_fold_perfs[0]:
                for k in [key for key in valid_fold_perfs[0] if key!='report']:vals=[p.get(k) for p in valid_fold_perfs if p.get(k) is not None and isinstance(p.get(k),(int,float))];
                if vals: avg_mets[f"avg_{k}"]=np.mean(vals)
                for met,val in avg_mets.items(): print(f"{met.replace('_',' ').title()}: {val:.4f}" if val else f"{met.replace('_',' ').title()}: N/A")
                valid_best_prms=[p for p in all_best_params_from_gs if p];
                if valid_best_prms:
                    param_s=[str(sorted(p.items())) for p in valid_best_prms]
                    if param_s: mc_param_s_tup=Counter(param_s).most_common(1)
                    if mc_param_s_tup: best_hyperparameters_overall=dict(eval(mc_param_s_tup[0][0]))
                if best_hyperparameters_overall: print(f"\nOverall Best HPs: {best_hyperparameters_overall}")
                else: print("Could not determine overall best HPs.")
                if all_feature_importances_outer:
                    try: concat_i=pd.concat(all_feature_importances_outer,axis=1).fillna(0);mean_i_s=concat_i.mean(axis=1).sort_values(ascending=False);print("\n--- Mean Feature Importances (Gain) (Top 3) ---");print(mean_i_s.head(3))
                    except Exception as e_imp: print(f"Error processing feature importances: {e_imp}")
            else: print("No valid fold perfs to aggregate.")
        else: print("Nested CV: No perfs recorded.")
        if best_hyperparameters_overall:
            print("\n--- Training Final Model ---");neg_t,pos_t=np.sum(y_full_labeled==0),np.sum(y_full_labeled==1);final_scale_w_val=neg_t/pos_t if pos_t*neg_t>0 else 1.0
            final_best_model=xgb.XGBClassifier(objective='binary:logistic',use_label_encoder=False,eval_metric='logloss',random_state=42,scale_pos_weight=final_scale_w_val,**best_hyperparameters_overall)
            final_best_model.fit(X_full_labeled,y_full_labeled);print("Final model trained.")
        else: print("Final model not trained: No best HPs.")
else: print("Skipping Nested CV: Initial data checks failed or N_OUTER_SPLITS < 2.")

# ================ 3. Evaluate Final Model on All Known Locations ================
training_data_with_final_preds = training_data.copy(); training_data_with_final_preds['Predicted_Label']=np.nan; training_data_with_final_preds['Predicted_Probability']=np.nan
if final_best_model:
    print("\n== 3. Evaluating Final Model on All Known Locations ==")
    # Pass through y_pred_proba_eval for ROC curve
    final_model_eval_metrics, y_pred_proba_eval_final = evaluate_model_metrics(
        final_best_model, 
        training_data[feature_cols].values, # Use values from training_data for X
        training_data['Label'].values,      # Use values from training_data for y
        "Final Model on Training Data"
    )
    # Update training_data_with_final_preds
    if final_model_eval_metrics: # Check if evaluation happened
        training_data_with_final_preds['Predicted_Label'] = final_best_model.predict(training_data[feature_cols].values)
        if y_pred_proba_eval_final is not None:
            training_data_with_final_preds['Predicted_Probability'] = y_pred_proba_eval_final
        
        # Print metrics
        print(f"\n--- Performance Metrics of Final Model on All Known Locations ---")
        for metric, value in final_model_eval_metrics.items():
            if metric != 'report': 
                 print(f"{metric.replace('_', ' ').title()}: {value if isinstance(value, str) else (f'{value:.4f}' if value is not None else 'N/A')}")
            else:
                print(f"\nClassification Report:\n{value}")

        # Plot Confusion Matrix
        if len(np.unique(training_data['Label'].values)) >= 2 and hasattr(final_best_model, 'classes_') :
            cm_final = confusion_matrix(training_data['Label'].values, training_data_with_final_preds['Predicted_Label'], labels=final_best_model.classes_)
            disp_final = ConfusionMatrixDisplay(confusion_matrix=cm_final, display_labels=final_best_model.classes_)
            disp_final.plot(cmap='Blues'); plt.title(f'Confusion Matrix - Final Model on All Known Locations'); plt.show()
        else:
            print("Skipping confusion matrix for known locations (single class or model classes issue).")

        # --- ADD ROC PLOT HERE ---
        if y_pred_proba_eval_final is not None and final_model_eval_metrics.get('roc_auc') is not None:
            fpr, tpr, _ = roc_curve(training_data['Label'].values, y_pred_proba_eval_final)
            roc_auc_val = final_model_eval_metrics['roc_auc']
            plt.figure(figsize=(8, 6))
            plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (AUC = {roc_auc_val:.2f})')
            plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
            plt.xlim([0.0, 1.0]); plt.ylim([0.0, 1.05])
            plt.xlabel('False Positive Rate'); plt.ylabel('True Positive Rate')
            plt.title('Receiver Operating Characteristic (ROC) - Final Model on Training Data')
            plt.legend(loc="lower right"); plt.grid(alpha=0.3); plt.show()
        else:
            print("Cannot plot ROC curve: Predicted probabilities or AUC not available for training data.")
        # --- END ROC PLOT ---
else: print("No final model to evaluate.")


# ================ 4. Predict Values on Unlabeled Data ================
print("\n== 4. Predicting on Unlabeled Data ==")
predictions_on_unlabeled_df = pd.DataFrame() # Initialize
if final_best_model and X_predict_unlabeled.size > 0:
    print(f"Making predictions for {X_predict_unlabeled.shape[0]} unlabeled points...")
    pred_proba_ul = final_best_model.predict_proba(X_predict_unlabeled)[:,1] # Probabilities for the positive class (geothermal)
    
    # --- APPLY CUSTOM THRESHOLD ---
    custom_threshold = 0.8 # Your desired threshold for higher precision
    print(f"Using custom threshold of {custom_threshold} for binary classification.")
    pred_bin_ul_custom_threshold = (pred_proba_ul >= custom_threshold).astype(int) # Converts boolean to 0 or 1
    # --- END CUSTOM THRESHOLD ---

    if not prediction_data.empty: 
        temp_pred_df = prediction_data.reset_index(drop=True) 
        predictions_on_unlabeled_df = pd.DataFrame({
            'Predicted_Probability': pred_proba_ul, 
            'Predicted_Label': pred_bin_ul_custom_threshold # USE THE NEW BINARY PREDICTIONS
        })
        
        cols_add = ['xcoord','ycoord','Area']
        for col_a in cols_add:
            if col_a in temp_pred_df.columns:
                if len(predictions_on_unlabeled_df)==len(temp_pred_df): 
                    predictions_on_unlabeled_df[col_a]=temp_pred_df[col_a].values
                else: 
                    print(f"Warning: Length mismatch for column '{col_a}' between predictions and original prediction_data. Skipping merge for this column.")
        
        print("\nPreview of predictions on unlabeled data (with custom threshold):")
        print(predictions_on_unlabeled_df.head()) 

        out_csv="XGB_Predict_Result_95.csv"
        try:
            predictions_on_unlabeled_df.to_csv(out_csv,index=False)
            print(f"\nPreds saved to: {out_csv}")
        except Exception as e:
            print(f"Error saving predictions to CSV: {e}")

        # --- MAPS FOR UNLABELED DATA PREDICTIONS WILL NOW REFLECT THE CUSTOM THRESHOLD ---
        if 'xcoord' in predictions_on_unlabeled_df.columns and 'ycoord' in predictions_on_unlabeled_df.columns:
            print("\n--- Mapping Predictions on Unlabeled Data (with custom threshold) ---")
            plt.figure(figsize=(12, 5))
            
            plt.subplot(1, 2, 1)
            scatter_binary = plt.scatter(
                predictions_on_unlabeled_df['xcoord'], 
                predictions_on_unlabeled_df['ycoord'], 
                c=predictions_on_unlabeled_df['Predicted_Label'], # This now uses the custom threshold
                cmap='coolwarm', s=5, alpha=0.7
            )
            plt.title(f'Binary Geothermal Predictions (Threshold = {custom_threshold})') # Update title
            plt.xlabel('xcoord'); plt.ylabel('ycoord')
            plt.colorbar(scatter_binary, ticks=[0, 1], label='Prediction (0=No, 1=Yes)')
            plt.grid(alpha=0.3)

            plt.subplot(1, 2, 2) # Probability map remains the same, as probabilities don't change
            scatter_proba = plt.scatter(
                predictions_on_unlabeled_df['xcoord'], 
                predictions_on_unlabeled_df['ycoord'], 
                c=predictions_on_unlabeled_df['Predicted_Probability'], 
                cmap='viridis', s=5, alpha=0.7, vmin=0, vmax=1
            )
            plt.title('Probability of Geothermal (Unlabeled Data)')
            plt.xlabel('xcoord'); plt.ylabel('ycoord')
            plt.colorbar(scatter_proba, label='Predicted Probability')
            plt.grid(alpha=0.3)
            
            plt.tight_layout(); plt.show()
        else:
            print("Coordinate columns ('xcoord', 'ycoord') not found in predictions_on_unlabeled_df. Cannot create maps.")
        # --- END MAPS ---
    else:
        print("Original prediction_data is empty, cannot add coordinate/area columns to predictions or map.")

elif not final_best_model: 
    print("No final model for preds.")
else: 
    print("No unlabeled data or X_predict_unlabeled is empty.")


# ================ 5. Further Performance Assessment ================
print("\n== 5. Further Performance Assessment ==")
if final_best_model and not training_data_with_final_preds.empty and \
   'Predicted_Label' in training_data_with_final_preds.columns and \
   not training_data_with_final_preds['Predicted_Label'].isnull().all():
    print("\n--- 5.1 Spatial Distribution of Errors ---")
    coord_cols_c = ['xcoord','ycoord']; x_c,y_c=None,None
    for col_c_cand in coord_cols_c: 
        if col_c_cand in training_data_with_final_preds.columns:
            if x_c is None: x_c=col_c_cand
            elif y_c is None and col_c_cand.lower()!=x_c.lower(): y_c=col_c_cand
    if x_c and y_c and 'Label' in training_data_with_final_preds.columns:
        tp_df=training_data_with_final_preds[(training_data_with_final_preds['Label']==1)&(training_data_with_final_preds['Predicted_Label']==1)]
        tn_df=training_data_with_final_preds[(training_data_with_final_preds['Label']==0)&(training_data_with_final_preds['Predicted_Label']==0)]
        fp_df=training_data_with_final_preds[(training_data_with_final_preds['Label']==0)&(training_data_with_final_preds['Predicted_Label']==1)]
        fn_df=training_data_with_final_preds[(training_data_with_final_preds['Label']==1)&(training_data_with_final_preds['Predicted_Label']==0)]
        plt.figure(figsize=(10,8));
        if not tn_df.empty: plt.scatter(tn_df[x_c],tn_df[y_c],c='blue',marker='.',s=10,alpha=0.3,label=f'TN ({len(tn_df)})')
        if not tp_df.empty: plt.scatter(tp_df[x_c],tp_df[y_c],c='green',marker='P',s=50,alpha=0.7,label=f'TP ({len(tp_df)})')
        if not fp_df.empty: plt.scatter(fp_df[x_c],fp_df[y_c],c='red',marker='X',s=50,alpha=0.7,label=f'FP ({len(fp_df)})')
        if not fn_df.empty: plt.scatter(fn_df[x_c],fn_df[y_c],c='orange',marker='s',s=30,alpha=0.7,label=f'FN ({len(fn_df)})') 
        plt.title('Spatial Errors (Known Data)');plt.xlabel(x_c);plt.ylabel(y_c);plt.legend();plt.grid(True,alpha=0.3);plt.show()
    else: print("Cannot create spatial error map: Coords or Labels missing.")

    print("\n--- 5.2 Partial Dependence Plots (PDP) ---")
    if hasattr(final_best_model,'feature_importances_') and X_full_labeled.size>0 and feature_cols and len(feature_cols)==X_full_labeled.shape[1]:
        imp_final=final_best_model.feature_importances_; sorted_idx_final=np.argsort(imp_final)[::-1]
        pdp_feat_idx=sorted_idx_final[:min(2,len(feature_cols))]
        if pdp_feat_idx.size>0:
            try:
                s_size_pdp=min(1000,X_full_labeled.shape[0]);pdp_s_idx=np.random.choice(X_full_labeled.shape[0],s_size_pdp,replace=False)
                X_pdp_s_for_check = pd.DataFrame(X_full_labeled[pdp_s_idx,:], columns=feature_cols)
                print("Dtypes for X_pdp_s (sample for PDP):"); X_pdp_s_for_check.info(verbose=False)
                PartialDependenceDisplay.from_estimator(final_best_model,X_pdp_s_for_check,features=pdp_feat_idx,feature_names=feature_cols,grid_resolution=50,kind='average')
                plt.suptitle("PDPs");plt.tight_layout(rect=[0,0,1,0.96]);plt.show()
            except Exception as e: print(f"PDP Error: {e}")
        else: print("No features for PDP.")
    else: print("PDP: Model lacks importances or data issues for PDP.")

    print("\n--- 5.3 SHAP Analysis (Summary Plot Only) ---")
    if X_full_labeled.size > 0 and feature_cols and len(feature_cols) == X_full_labeled.shape[1]: 
        try:
            explainer_shap = shap.TreeExplainer(final_best_model) 
            s_size_shap=min(200,X_full_labeled.shape[0]); shap_s_idx=np.random.choice(X_full_labeled.shape[0],s_size_shap,replace=False)
            X_shap_s_df = pd.DataFrame(X_full_labeled[shap_s_idx,:],columns=feature_cols) 
            print("Dtypes for X_shap_s_df (sample for SHAP summary):"); X_shap_s_df.info(verbose=False)
            shap_values_sample = explainer_shap.shap_values(X_shap_s_df)
            print("SHAP summary plot...");shap.summary_plot(shap_values_sample,X_shap_s_df,show=False);plt.title("SHAP Summary");plt.tight_layout();plt.show()
            # print("\nSHAP force plots section is commented out.")
        except ImportError: print("SHAP library not installed.")
        except Exception as e_s_main: print(f"SHAP summary plot analysis error: {e_s_main}") 
    else: print("Skipping SHAP Summary: X_full_labeled or feature_cols empty, or feature_cols length mismatch.")
else: print("Skipping further performance assessment: final model/predictions problematic.")
print("\n== End of Script ==")
```
